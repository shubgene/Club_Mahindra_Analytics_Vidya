```python
training_set.state_code_residence.fillna(8.0,inplace=True)
training_set.season_holidayed_code.fillna(2.0,inplace=True)
col_to_dummy=['channel_code','persontravellingid','resort_region_code','resort_type_code','state_code_resort','resort_id','state_code_residence','season_holidayed_code']

for col in col_to_dummy:
    df_pro=pd.get_dummies(training_set[col],prefix=col+'_')
    training_set=pd.concat([training_set,df_pro],axis=1)
    training_set.drop(columns=[col],inplace=True)
```

```python
training_set.head()
```

```python
training_set['days_stayed']=0
training_set['is_weekend']=0
training_set['month']=0
training_set['booking_checkin_diff']=0
training_set['Same_day_booking']=0
training_set['Advance_booking']=0
for index,row in training_set.iterrows():
    c_in=row['checkin_date']
    dt= datetime.datetime.now()
    dt_o= datetime.datetime.now()
    book_d=datetime.datetime.now()
    book_in=row['booking_date']
    try:
        book_d = datetime.datetime.strptime(book_in, '%d/%m/%y')
    except:
        book_d=datetime.datetime.strptime(book_in, '%d/%m/%Y')
    try:
        dt = datetime.datetime.strptime(c_in, '%d/%m/%y')
    except:
        dt=datetime.datetime.strptime(c_in, '%d/%m/%Y')
    c_out=row['checkout_date']
    try:
        dt_o = datetime.datetime.strptime(c_out, '%d/%m/%y')
    except:
        dt_o=datetime.datetime.strptime(c_out, '%d/%m/%Y')
    training_set.at[index,'month']=dt.month
    number_of_day=dt_o-dt
    diff=dt-book_d
    diff_d=diff.days
    training_set.at[index,'booking_checkin_diff']=diff_d
    if(diff_d==0):
        training_set.at[index,'Same_day_booking']=1
    if(diff_d>30):
        training_set.at[index,'Advance_booking']=1
    day=number_of_day.days
    training_set.at[index,'days_stayed']=day
    date1 = dt.toordinal()
    date2 = dt_o.toordinal()
    for i in range(date1, date2 + 1):
        dtc=datetime.date.fromordinal(i)
        if(dtc.weekday()>=5):
            training_set.at[index,'is_weekend']=1
            break
```
